```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "/content/drive/My Drive/Programacion/songs_normalize.csv"
```

```python
datos = pd.read_csv(data)
datos.head()
```

```python
nulos = datos.isnull().sum()
datos_clean = datos.dropna()
explicitas = datos_clean[datos_clean['explicit'] == True]
explicitas = explicitas.head(10)
```

```python
canciones = explicitas['artist'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
canciones.plot(kind='bar')
plt.title('Cantidad de canciones explícitas por artista')
plt.xlabel('Artista')
plt.ylabel('Cantidad de canciones explícitas')
plt.xticks(rotation=45)
plt.show()
```
